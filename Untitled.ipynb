```python
from qiskit import pulse, circuit
from qiskit.pulse import DriveChannel as d
```

```python
def some_function(param1, param2, param3):
    with pulse.build() as sched:
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param1))
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param2))
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param3))
    return sched

with pulse.build() as sched:
    pulse.call(some_function(0, 1, 2))
```

```python
@pulse.macro
def my_macro(param1, param2, param3):
    with pulse.align_right():
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param1))
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param2))
        pulse.play(pulse.Gaussian(100, 1.0, 20), d(param3))
    return param1*param2*param3
```

```python
with pulse.build():
    output = my_macro(1, 2, 3)
    print(output)
```

```python
@pulse.function()
def h():
    pulse.play(pulse.Gaussian(100, 1.0, 20), d(0))
    output = my_macro(1, 2, 3)
    print(output)
```
