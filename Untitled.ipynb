```python
@pulse.function()
def rz(qubit: int, theta: float):
    pulse.shift_phase(-theta, d(qubit))

@pulse.function()
def sx(qubit: int):
    pulse.play(pulse.Gaussian(100, 1.0, 10), d(qubit))

@pulse.function()
def u3(qubit: int, theta: float, phi: float, lamb: float):
    rz(qubit, phi+3*np.pi)
    sx(qubit)
    rz(qubit, theta+np.pi)
    sx(qubit)
    rz(qubit, lamb)
```

```python
@pulse.function()
def called():
    u3(2, 1, 2, 3)
    mem_slot = pulse.measure(2)
    return mem_slot
```
